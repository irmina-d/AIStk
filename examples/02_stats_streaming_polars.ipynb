# 02 — Streaming Stats with Polars Lazy
Compute per-MMSI statistics without materializing to NumPy.

```
import polars as pl
from aistk.core import AISDataset
from aistk.stats_streaming import compute_stats_lazy
```

```
ds = AISDataset('data/ais')
lf = ds._build()
res = compute_stats_lazy(lf, level='mmsi').collect(engine="streaming")
res.sort('distance_km', descending=True).head()
```
